```python
import numpy as np
import dask.array as da
from sklearn.cluster import KMeans as SklearnKMeans  # <-- use this only for weighted recluster (step 8)


try:
    import scipy.sparse as sp
    _HAVE_SP = True
except Exception:
    sp = None
    _HAVE_SP = False

def _ensure_dask_array(X):
    """Se X è SciPy sparse o NumPy, wrappalo in dask.array; altrimenti lascia X com'è."""
    if isinstance(X, da.Array):
        return X
    if _HAVE_SP and sp is not None and sp.issparse(X):
        # mantieni oggetto sparse nei blocchi; chunk per righe
        return da.from_array(X, asarray=False, chunks=(min(10000, X.shape[0]), X.shape[1]))
    # fallback per denso
    return da.from_array(X, chunks="auto")

def _row_norms_squared_block(block):
    """norma^2 per riga su blocco denso o SciPy sparse -> np.ndarray 1D."""
    if _HAVE_SP and sp is not None and sp.issparse(block):
        return np.asarray(block.multiply(block).sum(axis=1)).ravel()
    else:
        return np.einsum("ij,ij->i", block, block)

def _block_pairwise_sq_dists(block, C, C_norm2):
    """
    Distanze quadratiche fra righe del blocco (m x d) e centroidi C (k x d).
    Restituisce array denso (m x k).
    """
    if _HAVE_SP and sp is not None and sp.issparse(block):
        XC = (-2.0) * block.dot(C.T)       # (m,k)
    else:
        XC = (-2.0) * (block @ C.T)        # (m,k)

    rn2 = _row_norms_squared_block(block)  # (m,)
    XC += rn2[:, None]                     # aggiungi ||x||^2
    XC += C_norm2[None, :]                 # aggiungi ||c||^2
    return XC

def _pairwise_sq_dists(X, C):
    """
    Pairwise squared distances tra X (dask.array o compatibile) e C (np.ndarray denso).
    Ritorna dask.array denso (n_samples, k).
    """
    X = _ensure_dask_array(X)
    C = np.asarray(C)
    C_norm2 = np.einsum("ij,ij->i", C, C)  # (k,)
    return X.map_blocks(
        _block_pairwise_sq_dists,
        C, C_norm2,
        dtype=float,
        chunks=(X.chunks[0], (C.shape[0],))
    )

def _to_dense_row(x):
    """Converte un vettore riga (sparse o denso) in np.ndarray 1D denso."""
    if _HAVE_SP and sp is not None and sp.issparse(x):
        return np.asarray(x.toarray()).ravel()
    # Se arriva come np.matrix o shape (1, d), appiattisci
    return np.asarray(x).ravel()

# --- fine aggiunte minime ---


def phi(X,C):
    """
    cost function for a given set of centroids
    
    Input:
    - dataset X
    - set of centroids C
    Output:
    - Cost function
    """
    # sostituisce pairwise_distances(...).min(1).sum() con versione compatibile sparse
    return _pairwise_sq_dists(X, C).min(1).sum()

def assign_label(X,C):
    """
    Assign labels to each point of X corresponding to
    its closest centroid
    
    Input:
    - dataset X
    - set of centroids C
    Output:
    - set of labels label
    """
    # sostituisce pairwise_distances_argmin_min(...) con argmin su distanze quadre
    return da.argmin(_pairwise_sq_dists(X, C), axis=1)

def update_centroids(X,lab,k):
    """
    Function to update the position of the centroids
    
    Inputs:
    - dataset X
    - labels of the clusters lab
    - number of clusters k
    Output
    - New centroids C
    """
    # Versione reduce-based robusta per sparse: niente indicizzazione booleana su X.
    import dask.array as da
    from dask import delayed, compute

    # 🔧 Assicura che X sia un dask.array (gestisce anche SciPy sparse)
    X = _ensure_dask_array(X)

    # allinea i chunk delle etichette alle righe di X
    lab = da.asarray(lab).rechunk((X.chunks[0],))

    # blocchi ritardati allineati per righe
    x_blocks = X.to_delayed().ravel()
    l_blocks = lab.to_delayed().ravel()

    @delayed
    def _reduce_block(bx, bl):
        bl = np.asarray(bl)
        d = bx.shape[1]

        # 🔧 COO/CSC -> CSR per supportare bx[m] (boolean row indexing)
        if _HAVE_SP and sp is not None and sp.issparse(bx) and not sp.isspmatrix_csr(bx):
            bx = bx.tocsr()

        sums = np.zeros((k, d), dtype=float)
        counts = np.zeros((k,), dtype=np.int64)
        for i in range(k):
            m = (bl == i)
            if m.any():
                Xi = bx[m]  # locale al blocco: ok anche se è SciPy sparse CSR
                s = np.asarray(Xi.sum(axis=0)).ravel()  # funziona per denso e sparse
                sums[i] += s
                counts[i] += int(m.sum())
        return sums, counts

    parts = [_reduce_block(bx, bl) for bx, bl in zip(x_blocks, l_blocks)]
    results = compute(*parts)

    total_sums = sum(s for s, _ in results)
    total_counts = sum(c for _, c in results)

    # medie; i cluster vuoti resteranno NaN (gestiti in kmeans_parallel)
    C_new = np.divide(
        total_sums,
        total_counts[:, None],
        out=np.full_like(total_sums, np.nan, dtype=float),
        where=total_counts[:, None] != 0,
    )
    # ritorna come dask.array (compatibile con il resto del tuo codice)
    return da.from_array(C_new, chunks=(k, X.shape[1]))


def kmeans__init(X: da.Array, k: int, l: int, random_state=42):
    """
    k-means|| initialization (Dask)

    Parameters
    ----------
    X : dask.array, shape (n_samples, n_features)
    k : int
    l : int, oversampling factor

    Returns
    -------
    C : np.ndarray, shape (k, n_features)
        Initial centroids
    """
    rng = np.random.default_rng(random_state)

    X = _ensure_dask_array(X)  # (aggiunta minima per supportare input SciPy sparse)

    n = X.shape[0]
    # 1) pick first center
    i0 = rng.integers(n)
    c0 = X[i0].compute()
    C = _to_dense_row(c0)[None, :]

    # 2) initial cost
    psi = phi(X, C).compute()
    if psi <= 0:
        return C if len(C) == k else np.repeat(C, k, axis=0)[:k]
    
    # RandomState per riproducibilità
    rs = da.random.RandomState(random_state)
    
    # 3) O(log ψ) rounds
    t = max(1, int(np.ceil(np.log(psi))))
    for _ in range(t):
        d2min = _pairwise_sq_dists(X, C).min(1)
        p = da.clip(l * d2min / psi, 0.0, 1.0)
        # campionamento bernoulliano con RandomState
        samples = rs.random_sample(size=p.shape, chunks=p.chunks) < p
        # indici dei True
        mask = da.where(samples)[0].compute()
        mask = sorted(mask)   #https://github.com/dask/dask-ml/issues/39
        
        if np.any(mask):
            new_pts = X[mask].compute()
            if _HAVE_SP and sp is not None and sp.issparse(new_pts):
                new_pts = new_pts.toarray()
            C = np.vstack([C, np.asarray(new_pts)])
        psi = phi(X, C).compute()
        if psi <= 0:
            break

    # weights for k-means|| step 7
    dist2 = _pairwise_sq_dists(X, C)   # dask-friendly anche per sparse
    labels = da.argmin(dist2, axis=1)
    weights = da.bincount(labels, minlength=len(C)).compute().astype(float)

    # 8) recluster down to k  (usa SEMPRE scikit-learn qui)
    k = min(k, C.shape[0])
    km = SklearnKMeans(n_clusters=k, n_init=10, random_state=random_state)
    try:
        km.fit(C, sample_weight=weights)   # scikit-learn recente
    except TypeError:
        km.fit(C)                          # fallback per versioni vecchie
    return km.cluster_centers_


def kmeans_parallel(X, k, max_it=100, tol=1e-8, l=2):
    """
    Core function of our implementation: it implements kmeans with dask and kmeans|| initialization
    """
    C = kmeans__init(X, k, l) #initial centroids with k_means||
    for i in range(max_it): #iterative phase
        lab = assign_label(X, C)
        C_new = update_centroids(X, lab, k).compute()

        # gestisci cluster vuoti: sostituisci righe NaN col vecchio centro
        if np.isnan(C_new).any():
            nan_rows = np.any(np.isnan(C_new), axis=1)
            C_new[nan_rows] = C[nan_rows]

        if da.allclose(C, C_new, atol=tol):
            #print(f"Main KMeans Converged after {i+1} iterations.")
            break

        C = C_new

    return lab, C
```

```python
# --- test_kmeans_sparse_dask.py ------------------------------------------
import numpy as np
import scipy.sparse as sp
import dask.array as da

# Se le funzioni sono in un modulo separato, scommenta e adatta:
# from my_kmeans_module import kmeans_parallel, phi

# Altrimenti si assume che kmeans_parallel e phi siano già definite nel notebook.


def make_sparse_two_clusters(n_per_cluster=20, dim=2, sep=5.0, noise=0.2, seed=123):
    """Genera due cluster gaussiani in 2D e li restituisce come CSR e COO (SciPy)."""
    rng = np.random.default_rng(seed)
    A = rng.normal(loc=0.0, scale=noise, size=(n_per_cluster, dim))
    B = rng.normal(loc=sep,  scale=noise, size=(n_per_cluster, dim))
    X_dense = np.vstack([A, B])
    X_csr = sp.csr_matrix(X_dense)
    X_coo = X_csr.tocoo()
    return X_csr, X_coo


def to_dask_sparse(X, rows_per_chunk=None):
    """Wrappa una matrice SciPy sparse in dask.array mantenendo la sparsità dei blocchi."""
    n, d = X.shape
    if rows_per_chunk is None:
        rows_per_chunk = min(10000, n)   # chunk comodo per righe
    return da.from_array(X, asarray=False, chunks=(rows_per_chunk, d))


def run_once(X_da, k=2, l=2, max_it=50, tol=1e-6):
    """Esegue una singola corsa di kmeans_parallel su un dask.array (sparse o denso)."""
    lab, C = kmeans_parallel(X_da, k=k, max_it=max_it, tol=tol, l=l)
    lab_np = lab.compute()                         # materializza etichette
    cost = float(phi(X_da, C).compute())           # costo finale
    counts = np.bincount(lab_np, minlength=k)      # conteggi per cluster
    return lab_np, C, cost, counts


def main():
    print("== Test k-means|| con input SciPy sparse wrappati in Dask ==")
    X_csr, X_coo = make_sparse_two_clusters()

    # Wrap in Dask: CSR diretto; COO -> CSR PRIMA del wrap (evita problemi di slicing)
    X_csr_da = to_dask_sparse(X_csr)
    X_coo_da = to_dask_sparse(X_coo.tocsr())

    print("\n-- CSR run #1 --")
    lab1, C1, cost1, counts1 = run_once(X_csr_da)
    print("labels (prime 10):", lab1[:10])
    print("counts per cluster:", counts1)
    print("centroids:\n", C1)
    print("phi(X, C):", cost1)

    print("\n-- CSR run #2 (riproducibilità attesa) --")
    lab2, C2, cost2, counts2 = run_once(X_csr_da)
    print("labels (prime 10):", lab2[:10])
    print("counts per cluster:", counts2)
    print("centroids:\n", C2)
    print("phi(X, C):", cost2)

    same_labels = np.array_equal(lab1, lab2)
    same_centers = np.allclose(C1, C2, atol=1e-8)
    print("\nRiproducibilità (CSR): labels uguali?", same_labels, "| centri uguali?", same_centers)

    print("\n-- COO (convertito a CSR) run --")
    lab3, C3, cost3, counts3 = run_once(X_coo_da)
    print("labels (prime 10):", lab3[:10])
    print("counts per cluster:", counts3)
    print("centroids:\n", C3)
    print("phi(X, C):", cost3)

    # Check “di buon senso”: centri vicini a ~[0,0] e ~[5,5] (ordine arbitrario)
    approx_targets = np.array([[0.0, 0.0], [5.0, 5.0]])
    dists = np.linalg.norm(C1[:, None, :] - approx_targets[None, :, :], axis=2).min(axis=1)
    print("\nDistanza dei centri attesi (da ~[0,0] e ~[5,5]):", dists)


if __name__ == "__main__":
    main()
# -------------------------------------------------------------------------
```

```python
import numpy as np
import dask.array as da

try:
    import scipy.sparse as sp
    _HAVE_SP = True
except Exception:
    sp = None
    _HAVE_SP = False

def _l2_normalize_block(block, eps=1e-12):
    """
    Normalizza L2 per riga un blocco (denso o SciPy sparse).
    - Per sparse: usa diags(1/||x||) @ X per mantenere la sparsità.
    - Per denso: divide riga per riga.
    """
    if _HAVE_SP and sp is not None and sp.issparse(block):
        # ||x||_2 per riga
        norms = np.sqrt(block.multiply(block).sum(axis=1)).A1
        inv = 1.0 / np.maximum(norms, eps)
        D = sp.diags(inv)
        return D.dot(block)  # mantiene CSR/CSC
    else:
        # denso
        block = np.asarray(block)
        norms = np.linalg.norm(block, axis=1)
        inv = 1.0 / np.maximum(norms, eps)
        return (block.T * inv).T

def l2_normalize_dask(X, eps=1e-12):
    """
    Applica _l2_normalize_block a tutti i blocchi di X (mantiene i chunk).
    """
    X = X  # già dask.array
    return X.map_blocks(_l2_normalize_block, dtype=float)
```

```python
import numpy as np
import dask.array as da
import scipy.sparse as sp
import s3fs
import logging

logger = logging.getLogger(__name__)

def _is_sparse_npz(fs, path):
    """Ritorna True se `path` è un file npz salvato con scipy.sparse.save_npz."""
    try:
        # se load_npz riesce a caricare, è un npz sparse valido
        with fs.open(path, mode="rb") as f:
            _ = sp.load_npz(f)
        return True
    except Exception:
        return False

def _load_feature_shard(fs, path):
    import numpy as np
    import scipy.sparse as sp

    if path.endswith(".npz"):
        # Prova come sparse-npz (save_npz)
        try:
            with fs.open(path, "rb") as f:
                mat = sp.load_npz(f)
            # accetta solo 2D
            if mat.ndim == 2:
                return mat
        except Exception:
            pass
        # Fallback: npz generico con array densi
        with fs.open(path, "rb") as f:
            data = np.load(f, allow_pickle=True)
            # prova chiavi comuni
            for key in ("X", "features", "data"):
                if key in data and data[key].ndim == 2:
                    return data[key]
            # altrimenti prendi il primo array 2D
            for v in data.values():
                if isinstance(v, np.ndarray) and v.ndim == 2:
                    return v
        return None

    elif path.endswith(".npy"):
        with fs.open(path, "rb") as f:
            arr = np.load(f, allow_pickle=True)
        # scarta 1D (etichette)
        return arr if getattr(arr, "ndim", 1) == 2 else None

    else:
        raise ValueError(f"Unsupported shard format: {path}")


def load_dataset_from_cloudveneto(dataset_name, fs=None, bucket=None, row_chunk=20000):
    """
    Ritorna un dask.array (righe sharded) mantenendo i blocchi sparse quando possibile.
    """
    if fs is None:
        fs = create_s3_filesystem()
    if bucket is None:
        bucket = CLOUDVENETO_CONFIG["bucket"]

    prefix = f"{bucket}/datasets/{dataset_name}"
    files = fs.ls(prefix)

    shard_files = sorted([
        f for f in files
        if "shard_" in f and f.endswith((".npy", ".npz")) and "labels" not in f.lower() and "y" not in f.lower()
    ])

    if not shard_files:
        raise FileNotFoundError(f"No feature shards found for dataset {dataset_name}")

    parts = []
    n_rows_total = 0
    n_cols = None

    for path in shard_files:
        shard = _load_feature_shard(fs, path)
        if shard is None:
            continue

        # verifica dimensioni coerenti
        if n_cols is None:
            n_cols = shard.shape[1]
        elif shard.shape[1] != n_cols:
            raise ValueError(f"Shard {path} has {shard.shape[1]} cols, expected {n_cols}")

        # wrap in dask conservando sparse
        if sp.issparse(shard):
            chunks = (min(row_chunk, shard.shape[0]), shard.shape[1])
            Xp = da.from_array(shard, asarray=False, chunks=chunks)
        else:
            Xp = da.from_array(shard, chunks="auto")

        parts.append(Xp)
        n_rows_total += shard.shape[0]

    if not parts:
        raise ValueError("No valid 2D feature shards were loaded.")

    X = da.concatenate(parts, axis=0)
    logger.info(f"Loaded dataset '{dataset_name}' with shape={X.shape}, sparse_blocks={any(sp.issparse(p._meta) if hasattr(p,'_meta') else False for p in parts)}")
    return X

def run_distributed_kmeans(
    dataset_name: str,
    k: int = 10,
    l: int = 2,
    *,
    fs=None,
    bucket: str | None = None,
    row_chunk: int = 20000,
    persist_data: bool = True,
    normalize: str = "l2",   # "l2" (custom) oppure None
    verbose: bool = True
):
    """
    Carica dataset (anche sparse) da CloudVeneto, normalizza L2 per riga in modo sparse-safe,
    poi esegue k-means|| con Dask.
    """
    import numpy as np
    import dask.array as da
    import scipy.sparse as sp

    # 1) Filesystem e load (usa il tuo loader fixato)
    if fs is None:
        fs = create_s3_filesystem()
    X = load_dataset_from_cloudveneto(
        dataset_name, fs=fs, bucket=bucket, row_chunk=row_chunk
    )

    if verbose:
        try:
            is_sparse_meta = hasattr(X, "_meta") and sp.issparse(X._meta)
        except Exception:
            is_sparse_meta = False
        print(f"[load] X.shape={X.shape}, chunks={X.chunks}, sparse_blocks={is_sparse_meta}")

    # 2) Normalizzazione L2 (custom, sparse-safe)
    if normalize == "l2":
        Xn = l2_normalize_dask(X)
        if persist_data:
            Xn = Xn.persist()
        if verbose:
            print("[norm] L2 normalization applied (custom, sparse-safe).")
    elif normalize is None:
        Xn = X.persist() if persist_data else X
        if verbose:
            print("[norm] Skipped (using raw X).")
    else:
        raise ValueError("normalize deve essere 'l2' oppure None.")

    # 3) k-means|| (la tua implementazione già sparse-friendly)
    labels, centroids = kmeans_parallel(Xn, k=k, l=l)

    if verbose:
        print(f"[kmeans] centroids.shape={np.asarray(centroids).shape}")

    return labels, centroids
```
